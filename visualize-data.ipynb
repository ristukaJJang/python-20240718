##### 그래프의 종류
 ** 막대그래프 ** : 데이터를 카테고리별로 값을 비교하고자 할 때 사용하는 그래프 형식(수직 막대 그래프 or 수평 막대 그래프)
 > 카테고리별 데이터를 비교하고자 할 때 사용(제품 판매량, 지역별 인구수 )

 ** 선 그래프 ** : 시간의 흐름에 따라 데이터 포인터를 선으로 연결하여 사용하는 그래프 형식
 > 하나의 요소에 대해 시간의 흐름에 따라 변경값을 나타내고자 할 때 사용(기온, 주식 등)

 ** 파이 차트 ** : 전체 데이터에 대한 각 요소에 비율을 원형으로 나타낼 때 사용하는 그래프 형식
 > 전체에 대한 요소의 비율을 표현하고자할 때 사용

 ** 산점도 ** : 두 변수 간의 상관 관계를 표현할 때 사용하는 그래프 형식
 > 각 요소간 관계성을 파악할 때 사용(키와 몸무게의 상관 관계 비교)
 
 ** 히스토그램 ** : 범위별로 그룹화하여 각 그룹에 속하는 데이터의 수를 막대로 표현하는 그래프 형식
 > 데이터의 분포를 비교할 때 사용(시험 점수 분포)

 ** 영역 그래프 * : 선 그래프의 아래쪽 빈 공간을 채운 그래프
 > 시간의 흐름에 따라 데이터를 비교할 때 사용하는데, 변화량을 강조하고자 할 때 사용

 ** 박스 그래프 ** : 데이터의 분포를 나타날 때 사용하는 그래프 형식
 > 최대값과 최소값 그리고 사분위수로 데이터의 이상치를 판단할 때 사용



### 선 그래프
가장 기본적인 그래프 형식, 시간의 흐름 및 순서에 따라 데이터의 변화량을 보여주고자 할 때 사용
plot 패키지의 plot() 메서드로 간단하게 생성가능

````python
plt.plot([1,2,4,8,16])
plt.show()
````

````python
# 그래프의 제목 지정 title() 메서드 사용
# plot(x축에 대한 리스트, y축에 대한 리스트)
plt.title('x축 추가')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16])
plt.show()
````

````python
# 차트 스타일 지정
# plot() 메서드의 세번째 매개변수로 지정된 문자열 포맷으로 전달하여 스타일 지정가능
# 스타일 지정 가능
plt.title('c2-- 스타일 적용')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], 'c2--')
plt.show()
````

````python
# 스타일 매개변수
# color(c) : 선 색상, linewidth(lw) : 선 굵기, linestyle(ls) : 선 스타일
# marker : 마커 종류, markersize(ms) : 마커 크기, markeredgecolor(mec) : 마커 테두리 색상
# markerfacecolor(mfc) : 마커 내부 색상, markeredgewidth(mew) : 마커 테두리 너비
plt.title('c2-- 스타일 적용')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], c = '#00998850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#009988")
plt.show()
````

````python
# 차트 범위 지정
# xlim(), ylim() 메서드로 x축과 y축의 상하한을 지정할 수 있음
plt.title('차트 범위 지정')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], c = '#00998850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#009988")
plt.xlim(2019, 2025)
plt.ylim(3, 20)
plt.show()
````

````python
# 틱 지정 및 그리드 지정
# xticks(), yticks() 메서드로 틱의 위치나 라벨 지정 가능
plt.title('차트 범위 지정')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], c = '#00998850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#009988")
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.grid(True)
plt.show()
````

````python
# 겹쳐 그리기
# xticks(), yticks() 메서드로 틱의 위치나 라벨 지정 가능
plt.title('겹쳐그리기')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], c = '#00998850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#009988")
plt.plot([2024, 2023, 2022, 2021, 2020], [1, 2, 4, 8, 16], c = '#991850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#991850")
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.grid(True)
plt.show()
````

````python
# 범례 표시
# legend() 메서드를 통해 범례를 표시
# loc 매개변수에 위치에 해당하는 정수를 전달하여 위치 지정 가능
plt.title('범례 표시')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], c = '#00998850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#009988", label = "증가")
plt.plot([2024, 2023, 2022, 2021, 2020], [1, 2, 4, 8, 16], c = '#991850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#991850", label = "감소")
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.grid(True)
plt.legend(loc = 9)
plt.show()
````

````python
# 각 틱 라벨
# xlabel(), ylabel() 메서드로 x축, y축 라벨 표시 가능
# loc 매개변수에 위치에 해당하는 정수를 전달하여 위치 지정 가능
plt.title('범례 표시')
plt.plot([2020, 2021, 2022, 2023, 2024], [1, 2, 4, 8, 16], c = '#00998850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#009988", label = "증가")
plt.plot([2024, 2023, 2022, 2021, 2020], [1, 2, 4, 8, 16], c = '#991850', lw = 5, ls=':', marker='*',
         ms = 10, mfc = "#991850", label = "감소")
plt.xticks([2020, 2022, 2024])
plt.yticks([0, 8, 16], ['저점', '중간점', '고점'])
plt.xlabel('연도')
plt.ylabel('가치 ')
plt.grid(True)
plt.legend(loc = 9)
plt.show()
````

### 막대 그래프
x축의 값이 분류가 가능한 카테고리 형태일 때 bar(), barh() 메서드로 막대 그래프 표현

````python
plt.bar(['a', 'b', 'c', 'd'],[13, 93, 28, 27])
plt.show()
````

````python
#가로 막대그래프 : barh()
plt.title('막대 그래프')
plt.barh(['a', 'b', 'c', 'd'],[13, 93, 28, 27], color='#ff0159')
plt.show()
````

````python
#가로 막대그래프 : barh()
# 오차 표시 : xerr, yerr 매개변수로 오차 지정 가능
plt.title('막대 그래프')
plt.barh(['a', 'b', 'c', 'd'],[13, 93, 28, 27], color='#ff000050', xerr=[2, 10, 4, 1])
plt.show()
````

### 스템 플롯
기준점으로부터 오차 범위를 표현하고자 할 때 사용

````python
x = np.linspace(0.1, 2* np.pi, 10)
plt.title('스템 플롯')
plt.stem(x, np.cos(x))
plt.show()
````

### 파이차트
각 분류가 전체에서 차지하는 비중을 표현하고자 할 때 사용

````python
plt.title('파이 차트')
plt.pie([25, 45, 39], labels=['사과', '바나나', '포도'], explode = [0, 0.1, 0], colors=['red', 'yellow', 'green'], shadow=True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
````

### 히스토그램
특정 구간에 대한 집계 데이터를 표시할 때 사용하는 그래

````python
x = np.random.randn(1000)
plt.title('히스토그램')
plt.hist(x, bins=10)
````

### 점 그래프
변수 간의 상관관계 표현할 때 사용하는 그래프

````python
x = np.random.normal(0, 1, 100)
y1 = np.random.normal(0, 1, 100)
y2 = np.random.normal(0, 1, 100)

plt.title('점 그래프')
plt.scatter(x, y1)
plt.grid(True)
plt.show()
````

````python

plt.title('버블 차트')
plt.scatter(x, y1, c=y2)
plt.grid(True)
plt.show()
````

### 컨투어 플롯
3차원 자료에서 x, y 입력과 z출력이 존재할 때, z를 선과 색상으로 표현하는 그래프입니다

````python
def func(x, y):
    z = (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
    return z

x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
xx, yy = np.meshgrid(x, y)
zz = func(xx, yy)

plt.title('등고선 그래프')
plt.contour(xx, yy, zz)
````
